```python
import math
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

from scipy.misc import derivative
```

```python
@np.vectorize
def func(x, y):
    
    return x**2 + y**2 + x*y
```

```python
plt.figure(figsize = (14, 8))

ax = plt.axes(projection = '3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x,y)

Z = func(X, Y)

ax.plot_surface(X, Y, Z, color = 'purple', antialiased = True, edgecolor = 'purple') 
plt.xlabel('X-axis') 
plt.ylabel('Y-axis') 
plt.grid(True)
```

```python
def get_partial_derivative(func, const_value, x=None, y=None):
    h = 0.0001
    
    if x is not None:
        y = const_value
        return(func(x + h, y) - func(x,y)) / h
    
    if y is not None:
        x = const_value
        return (func(x, y + h) - func(x, y)) / h
```

```python
const_value = 1

const_value
```

```python
get_partial_derivative(func, const_value, x=1)
```

```python
get_partial_derivative(func, const_value, x=5)
```
